```python
import json
```

```python
with open("cards_Database.json","r") as file:
    cards = json.load(file)
```

```python
cards
```

```python
subtypes = []
ninjas = []
for i in cards:
    try:
        for j in i["Subtipos"]:
            if j not in subtypes:
                subtypes.append(j)
            if j == "Ninja" and i not in ninjas:
                ninjas.append(i)
    except:
        pass
```

```python
subtypes
```

```python
ninjas
```

```python
with open("ninjas.json","w") as file:
    a = json.dumps(ninjas)
    file.writelines(a)
```
