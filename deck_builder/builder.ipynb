Explicação sobre a técnica de comparação utilizada, a distância de Jaccard:

O Jaccard Distance é uma medida de dissimilaridade entre dois conjuntos. É uma maneira de medir a similaridade entre dois conjuntos, definida como o tamanho da interseção dividido pelo tamanho da união dos conjuntos.

A fórmula para calcular a distância de Jaccard é a seguinte:
J(A, B) = 1 - |A ∩ B| / |A ∪ B|

A distância de Jaccard é uma métrica de dissimilaridade, o que significa que quanto mais próximos os conjuntos, menor será a distância de Jaccard. Se os conjuntos forem idênticos, a distância de Jaccard será 0. Se os conjuntos não tiverem nenhum elemento em comum, a distância de Jaccard será 1.

Na prática, a distância de Jaccard é frequentemente usada para medir a similaridade entre conjuntos de elementos, como palavras em documentos de texto, n-gramas em sequências de caracteres ou itens em conjuntos de dados. É amplamente utilizada em tarefas de mineração de texto, recuperação de informação, classificação e agrupamento de dados.

```python
from nltk.util import ngrams
from nltk.metrics import jaccard_distance
```

```python
def calcular_similaridade(texto1, texto2, n=1):
    shingles1 = set()
    shingles2 = set()
    
    # Crie shingles para o primeiro texto
    for i in range(len(texto1) - n + 1):
        shingle = texto1[i:i+n].lower()
        shingles1.add(shingle)
    
    # Crie shingles para o segundo texto
    for i in range(len(texto2) - n + 1):
        shingle = texto2[i:i+n].lower()
        shingles2.add(shingle)
    
    # Calcule a similaridade de Jaccard entre os shingles
    intersection = len(shingles1.intersection(shingles2))
    union = len(shingles1.union(shingles2))
    
    return intersection / union if union > 0 else 0

def buscar_carta_p_texto(cartas, texto_consulta, limiar=0.3, ngram_size=2):
    resultados = []
    texto_consulta = texto_consulta.lower()  # Normaliza o texto de consulta
    
    for carta in cartas:
        try:
            texto_carta = carta.get('Texto', '').lower()  # Normaliza o texto da carta
            similaridade = calcular_similaridade(texto_consulta, texto_carta, n=ngram_size)
            if similaridade >= limiar:
                resultados.append((carta, similaridade))
        except Exception as e:
            pass
    
    # Ordena os resultados com base na similaridade
    resultados.sort(key=lambda x: x[1], reverse=True)
    
    return resultados
```

```python
# Realiza a busca
entrada = "creatures you control get"

calcular_parametros = lambda entrada: (0.3, 3) if len(entrada.split()) > 3 else ((0.2, 2) if len(entrada.split()) > 1 else (0.157, 2))

limiar, ngram_size = calcular_parametros(entrada)

print(limiar,ngram_size)

resultados = buscar_carta_p_texto(cards, entrada, limiar=limiar, ngram_size=ngram_size)
```

```python
cards_result = {}
# Exibe os resultados
for carta, similaridade in resultados:
    if carta["Nome"] not in cards_result.keys():
        cards_result[carta["Nome"]] = carta
```
